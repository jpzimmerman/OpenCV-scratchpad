```python
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

video_output = cv2.VideoWriter("./mars_high_output.mp4", cv2.VideoWriter_fourcc('M','P','4','2'), 60, (frame_width, frame_height))
```

```python
while (cap.isOpened()):
    ret, frame = cap.read()

    if ret:
        video_output.write(frame)
    else: break;

video_output.release()
```
