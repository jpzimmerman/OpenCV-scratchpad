```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
base_image = cv2.imread("./summer-landscape.jpg", cv2.IMREAD_GRAYSCALE)
rgb_formatted_image = cv2.cvtColor(base_image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_formatted_image)
```

```python
img = cv2.line(rgb_formatted_image, (100,100), (300,300), (0,255,0), lineType=cv2.LINE_16)
```

```python
plt.imshow(img)
```

```python
img = cv2.circle(rgb_formatted_image, (100,100), 50, (0,0,255), thickness=-10, lineType=cv2.LINE_AA)
```

```python
plt.imshow(rgb_formatted_image)
```

```python
img = cv2.rectangle(rgb_formatted_image, (300,50), (400,150), (255,0,0), thickness=5, lineType=cv2.LINE_AA)
```

```python
plt.imshow(rgb_formatted_image)
```

```python
img = cv2.putText(rgb_formatted_image, "Lake Scene (Annotated)", (100, 320), cv2.FONT_HERSHEY_COMPLEX, -1.1, (255,200,60), 2, cv2.LINE_AA)
```
